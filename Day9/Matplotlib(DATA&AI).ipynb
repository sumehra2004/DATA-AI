```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv("stud.csv")
print("Original Data:\n", df, "\n")

print("Missing values:\n", df.isnull(), "\n")

# Fill missing Age and Marks using mean
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Marks"].fillna(df["Marks"].mean(), inplace=True)

# Clean City column (capitalize properly)
df["City"] = df["City"].str.strip().str.title()

print("Cleaned Data:\n", df, "\n")

# Analysis
# Average marks
avg_marks = df["Marks"].mean()
print("Average Marks:", avg_marks)

# Top scoring student
top_student = df.loc[df["Marks"].idxmax()]
print("\nTop Scoring Student:\n", top_student)

# Students scoring below 70
below_70 = df[df["Marks"] < 70]
print("\nStudents scoring below 70:\n", below_70)

# Visualization
plt.bar(df["Name"], df["Marks"])
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Marks vs Students")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Raw data
data = {
    "Day": ["mon", "tue"],
    "Product": ["laptop", "phone"],
    "Price": [800, 500],
    "Quantity": ["2", "three"]
}

df = pd.DataFrame(data)

# Data cleaning
df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce")
df["Quantity"] = df["Quantity"].fillna(0)

# Create total_sales
df["Total_Sales"] = df["Price"] * df["Quantity"]

# Total sales per product
sales_per_product = df.groupby("Product")["Total_Sales"].sum()

# Best selling product
best_product = sales_per_product.idxmax()

print("Total Sales per Product:")
print(sales_per_product)
print("\nBest Selling Product:", best_product)

# Visualization
plt.figure()
plt.bar(df["Day"], df["Total_Sales"])
plt.xlabel("Day")
plt.ylabel("Total Sales")
plt.title("Total Sales by Day")
plt.show()
```
